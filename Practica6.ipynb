````python

# Configuration and paramters

df_titanic = read_csv(input_path)

nan_values_by_column = df_titanic.isna().sum()

print('Show Nan data \n')
print(df_titanic.isna().sum())
print(f'Total registers:{df_titanic.index.size}')
print('\n')

df_titanic
````

### Ejercicio 1: Análisis de la distribución de supervivencia por combinación de sexo y clase del pasajero.

- Calcula la proporción de supervivencia para cada combinación de 'Sex' y 'Pclass'.
- Identifica qué combinación tuvo la tasa de supervivencia más alta.
- Identifica qué combinación tuvo la tasa de supervivencia más baja.


````python
df_titanic['Pclass'] = df_titanic['Pclass'].map({1: "first", 2:"second", 3:"third"})

sex_pclass_group = df_titanic.groupby(['Sex','Pclass'])['Survived'].mean()

sex_pclass_df = sex_pclass_group.reset_index()

sex_pclass_df['Survived'] = sex_pclass_df[['Survived']].apply(lambda x : x * 100)

table_grouped = tabulate.tabulate(sex_pclass_df, headers=["Sex", "Class", "Survived Rate %"], tablefmt='github', floatfmt=".2f", showindex=False)

max_group = sex_pclass_group.idxmax()
max_value = sex_pclass_group.max() 
min_group = sex_pclass_group.idxmin()
min_value = sex_pclass_group.min() 

print(table_grouped + '\n')
print(f'Max survivors group: {max_group} {max_value*100:.2f} % \n')
print(f'Min survivors group: {min_group} {min_value*100:.2f} % \n')
````

### Ejercicio 2: Identificación de familias grandes a bordo.

- Crea una nueva columna ```FamilySize``` sumando las columnas ```SibSp``` y ```Parch```.
- Considera como ```familia grande``` a aquellas donde ```FamilySize``` es mayor a 3.
- Calcula el número de pasajeros en familias grandes.
- Calcula la proporción de supervivencia entre los pasajeros que pertenecen a familias grandes.


````python
df_titanic['FamiliSize'] = df_titanic['Parch'] + df_titanic['SibSp']

is_big_families = df_titanic['FamiliSize']  > 3

df_big_family= df_titanic

df_big_family['is_bigfamily'] = is_big_families

df_big_family['is_bigfamily'] = df_big_family['is_bigfamily'].map({True: 'Yes', False: 'No'})

is_big_fam_series = df_big_family.groupby('is_bigfamily').size()

is_big_families_proportion = df_big_family.groupby('is_bigfamily')['Survived'].mean()

print(f'Pasajeros en familias grandes: {is_big_fam_series["Yes"]}')

print(f'De los pasajeros de familias grandes {is_big_families_proportion["Yes"]*100:.2f} % sobrevivieron')

````

### Ejercicio 3: Segmentación por grupos de edad.

Clasifica a los pasajeros en las siguientes categorías de edad (tip puede resultar mas sencillo realizarlo con una función)

- Menor de Edad, cuando tenga menos de 18 años
- Mayor de Edad, cuando tenga 18 años o más

````python
def has_age_mayirity(age):
    if (np.isnan(age)): return np.nan
    return age >= 18

df_titanic['Age_of_mayority'] = df_titanic['Age'].apply(has_age_mayirity)

has_age_mayority_series = df_titanic.groupby('Age_of_mayority').size()

print(f'Found {has_age_mayority_series.iloc[1]} that has age of mayority and {has_age_mayority_series.iloc[0]} that not.')
print(f'NaN values was {nan_values_by_column.loc["Age"]}.')
````

### Ejercicio 4: Comparación entre promedios calculados manualmente y con Pandas

### Ejercicio 5. Creación de intervalos de clase usando NumPy y análisis con Pandas

- Divide la columna ```Fare``` en 5 intervalos equidistantes utilizando la función numpy.linspace, el estudiante deberá investigar la operación de esta función en python.
- Crea una nueva columna en el DataFrame que asigne a cada pasajero el intervalo correspondiente de su tarifa.
- Calcula el número de pasajeros en cada intervalo utilizando Pandas y la proporción de supervivientes por intervalo.

````python
fare_min = df_titanic['Fare'].min()
fare_max = df_titanic['Fare'].max()

interval_fare = np.linspace(fare_min,fare_max, 6)

df_titanic['Fare_class'] = pd.cut(df_titanic['Fare'], bins=interval_fare, include_lowest=True)

df_interval = df_titanic.groupby('Fare_class',observed=False).size().reset_index()

df_interval['Survived'] = df_interval['Fare_class'].map(df_titanic.groupby('Fare_class',observed=False)['Survived'].mean())

df_interval['Survived'] = df_interval['Survived'].apply(lambda x: 0 if np.isnan(x) else x)

inerval_table = tabulate.tabulate(df_interval
                                  , showindex=False, tablefmt='github', floatfmt='.2%'
                                  , headers=['Fare class', 'Frecuency', 'Proportion survived'])
print('Intervale table')
print(inerval_table)

````
